```python
from keras import backend as K

def root_mean_squared_error(y_true, y_pred):
        return K.sqrt(K.mean(K.square(y_pred - y_true), axis=-1)) 
```

```python
from keras.layers import Add, Activation, Lambda
from keras.models import Model
from keras.layers import Input, Reshape, Dot
from keras.layers.embeddings import Embedding
from keras.optimizers import Adam
from keras.regularizers import l2
import keras.losses

class Embed:
    def __init__(self, n_items, n_factors):
        self.n_items = n_items
        self.n_factors = n_factors
    
    def __call__(self, x):
        x = Embedding(self.n_items, self.n_factors, embeddings_initializer='he_normal', embeddings_regularizer=l2(1e-6))(x)
        x = Reshape((self.n_factors,))(x)
        
        return x
    
def NCF(n_users, n_rests, n_factors, min_rating, max_rating):
    user = Input(shape=(1,))
    usr = Embed(n_users, n_factors)(user)
    bias_u = Embed(n_users, 1)(user)
    
    r = Input(shape=(1,))
    itm = Embed(n_rests, n_factors)(r)
    bias_m = Embed(n_rests, 1)(r)   
    
    x = Dot(axes=1)([usr, itm])
    x = Add()([x, bias_u, bias_m])
    x = Activation('sigmoid')(x)
    x = Lambda(lambda x: x * (max_rating - min_rating) + min_rating)(x)  
    
    model = Model(inputs=[user, r], outputs=x)
    opt = Adam(lr=0.001)
    model.compile(loss = 'mean_squared_error',optimizer=opt, metrics=['mse', 'mae', 'mape'])  
    
    return model
```

```python
keras_model = NCF(n_users, n_rests, n_factors, min_rating, max_rating)
keras_model.summary()
```
